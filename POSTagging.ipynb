## Named Entity Recognition

#### NER is the form of NLP.

```python

import spacy
```

```python

nlp=spacy.load('en_core_web_sm')
```

```python

import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
def show_entities(d):
    if d.ents:
        for ent in d.ents:
            print(f"{ent.text:{15}}{ent.label_:{15}}{spacy.explain(ent.label_)}")
    else:
        print("No entities found ")
```

```python
d=nlp(u"GFG is an Indian company which provides one of the finest education")
```

```python

show_entities(d)
```

```python

d=nlp(u"I am not feeling well today")
```

```python

show_entities(d)
```

## How would we add new entity

```python
d=nlp(u"Tesla is earning money at an extensive rate")
```

```python

show_entities(d)
```

```python
from spacy.tokens import Span as ss
```

```python

ORG=d.vocab.strings[u"ORG"]
```

```python
new_entity=ss(d,0,1,label=ORG)
#0,1 is the index position of Tesla in the text
```

```python
d.ents=list(d.ents)+[new_entity]
#Here we are adding it to our entities list
```

```python
d
```

```python

show_entities(d)
```

## Lets see how to add multiple words as an entity

```python
d=nlp(u"Playing cricket and football are both good for health")
```

```python

show_entities(d)
```

```python
from spacy.matcher import PhraseMatcher
```

```python

m=PhraseMatcher(nlp.vocab)
```

```python
phrase=['football','cricket']
```

```python
patterns=[nlp(text) for text in phrase]
#This is a list Comphrehension
```

```python
m.add('sports',None,*patterns)
#Press Shift Tab to check the parameters
```

```python
found=m(d)
```

```python
found
```

```python
from spacy.tokens import Span as ss
```

```python
d
```

```python

sport=d.vocab.strings[u"Sports"]
```

```python
new_ents=[ss(d,match[1],match[2],label=sport) for match in found]
```

```python

d.ents=list(d.ents)+new_ents
```

```python

d
```

```python

show_entities(d)
```

# Word Cloud

```python
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
```

```python
corpus='''India, officially the Republic of India (Hindi: Bhārat Gaṇarājya),[25] is a country in South Asia. It is the seventh-largest country by area, the second-most populous country, and the most populous democracy in the world. Bounded by the Indian Ocean on the south, the Arabian Sea on the southwest, and the Bay of Bengal on the southeast, it shares land borders with Pakistan to the west;[f] China, Nepal, and Bhutan to the north; and Bangladesh and Myanmar to the east. In the Indian Ocean, India is in the vicinity of Sri Lanka and the Maldives; its Andaman and Nicobar Islands share a maritime border with Thailand, Myanmar, and Indonesia.'''
```

```python
corpus = corpus.replace("[25]" , "")    
corpus = corpus.replace("[f]" , "")   
corpus = corpus.replace(")" , "")
```

```python

print(corpus)
```

### stop word removel

```python
words=[]
for word in word_tokenize(corpus):
    if (word.lower() not in stopwords.words('english')) and (len(word)>=2):
        words.append(word.lower())
```

#### creating vocab

```python
vocab=list(set(words))
len(vocab)
```

#### Create Encoders and Decoders

```python
num=1
word_to_num={}
num_to_word={}
for word in vocab:
    word_to_num[word]=num
    num_to_word[num]=word
    num+=1
```
